```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

```python
tss = []
data_sets = ["adalt", "car", "white", "red", "google", "trip"]
data_sets_n = ["adult", "car", "white", "red", "google", "trip"]

for ts in data_sets: 
    wgs = []
    for wg in [1]:
        eps = [] 
        for ep in [2,3,4,5,6]:
            eps.append(np.loadtxt('result/ral_multi_class_log_{ts}_episode_{ep}_weight_{wg}.csv'.format(ts=ts, ep=ep, wg=wg),
                                  delimiter='\n'))
        wgs.append(np.mean(np.asarray(eps),axis=0))
    tss.append(np.asarray(wgs))
tss = np.asarray(tss)
```
